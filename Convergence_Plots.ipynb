```python
import pandas as pd

# Load the input CSV file
input_file = "0-99.csv"

# Initialize variables
output_data = {'Number of Images': [], 'captions': []}
current_captions = []

# Read the file line by line
with open(input_file, 'r') as file:
    for line in file:
        line = line.strip()

        # Skip lines enclosed in double quotes
        if line.startswith('"') and line.endswith('"'):
            continue

        # Check if the line contains "Image" and process accordingly
        if line.startswith('Image'):
            # Start a new ID
            current_captions = [caption.strip() for caption in line.split(':')[1:]]
        elif line.startswith('Error'):
            current_captions.append("Error while doing image captioning.")
        elif line:
            current_captions.extend([caption.strip() for caption in line.split(':')])

        # If it's a new ID or the last line, save the previous data
        if line.startswith('Image') or not line:
            # Remove the pattern "Image <number of next image>" from the last sentence (except the last one)
            for i in range(len(current_captions) - 1):
                current_captions[i] = current_captions[i].rsplit('Image', 1)[0].strip()

            output_data['Number of Images'].append(len(current_captions))
            output_data['captions'].append(current_captions)

# Create a DataFrame from the collected data
output_df = pd.DataFrame(output_data)

# Save the output DataFrame to a new CSV file
output_file = "op.csv"
output_df.to_csv(output_file, index=False)

print(f"Output CSV file '{output_file}' created successfully.")
```
